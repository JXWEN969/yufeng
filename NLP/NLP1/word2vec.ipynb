```python
import gensim
import logging
import multiprocessing
from gensim.models import word2vec
import numpy as np
from time import time
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

data_path = r'D:\\Word2VecTransE(1)\Word2VecTransE\Word2Vec\text8'
sents = word2vec.Text8Corpus(data_path)

# 
def evaluate_model(model):
    sims = []
    ground_truth = []
    with open(r'D:\\Word2VecTransE(1)\Word2VecTransE\Word2Vec\wordsim353\combined.csv') as f:
        for line in f.readlines()[1:]:
            l = line.strip().split(',')
            if l[0] in model.wv.key_to_index and l[1] in model.wv.key_to_index:
                sims.append(model.wv.similarity(l[0], l[1]))
                ground_truth.append(float(l[2]))
    spcor = spearmanr(ground_truth, sims)[0]
    return spcor

def train_and_evaluate(num_rounds=10):
    vector_size = 150
    window_size = 20
    min_count = 30
    scores = []

    for round in range(num_rounds):
        print(f"Training round {round+1}/{num_rounds}")
        model = gensim.models.Word2Vec(sents, vector_size=vector_size, window=window_size, 
                                       min_count=min_count, workers=multiprocessing.cpu_count())
        score = evaluate_model(model)
        scores.append(score)
        print(f"Round {round+1}: Score {score}")

    # 
    best_score = max(scores)
    best_round = scores.index(best_score) + 1
    print(f"The best score is {best_score}, achieved in round {best_round}")

    return scores

# 1-10
def plot_scores_by_round(scores, num_rounds=10):
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, num_rounds+1), scores, 'o-', color='blue')
    plt.title('Scores by Round')
    plt.xlabel('Round Number')
    plt.ylabel('Score')
    plt.xticks(range(1, num_rounds+1))
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    scores = train_and_evaluate()
    plot_scores_by_round(scores)
```

```python
import gensim
import logging
import multiprocessing
from gensim.models import word2vec
import numpy as np
from time import time
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

data_path = r'D:\\Word2VecTransE(1)\Word2VecTransE\Word2Vec\text8'
sents = word2vec.Text8Corpus(data_path)

def evaluate_model(model):
    sims = []
    ground_truth = []
    with open(r'D:\\Word2VecTransE(1)\Word2VecTransE\Word2Vec\wordsim353\combined.csv') as f:
        for line in f.readlines()[1:]:
            l = line.strip().split(',')
            if l[0] in model.wv.key_to_index and l[1] in model.wv.key_to_index:
                sims.append(model.wv.similarity(l[0], l[1]))
                ground_truth.append(float(l[2]))
    spcor = spearmanr(ground_truth, sims)[0]
    return spcor

def train_and_evaluate_by_config():
    configurations = [
        {"vector_size": 200, "window": 10, "min_count": 10, "sg": 1, "epochs": 5, "suffix": "4_1"},
        {"vector_size": 200, "window": 10, "min_count": 10, "sg": 1, "epochs": 10, "suffix": "4_2"},
        {"vector_size": 200, "window": 10, "min_count": 10, "sg": 1, "epochs": 15, "suffix": "4_3"}
    ]
    results = []

    for config in configurations:
        model = gensim.models.Word2Vec(sents, vector_size=config['vector_size'], window=config['window'],
                                       min_count=config['min_count'], sg=config['sg'], workers=multiprocessing.cpu_count())
        model.train(sents, total_examples=model.corpus_count, epochs=config['epochs'])
        score = evaluate_model(model)
        results.append((config['suffix'], score))
        print(f"Config Suffix {config['suffix']}: Score {score}")

    return results

# 
def plot_scores_by_config(results):
    plt.figure(figsize=(10, 5))
    suffixes = [res[0] for res in results]
    scores = [res[1] for res in results]
    plt.bar(suffixes, scores, color='blue')
    plt.title('Configuration Suffix vs Score')
    plt.xlabel('Configuration Suffix')
    plt.ylabel('Score')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    results = train_and_evaluate_by_config()
    plot_scores_by_config(results)
```
